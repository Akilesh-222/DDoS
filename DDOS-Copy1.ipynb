```python
# import the necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset from the csv file using pandas
data = pd.read_csv(r"C:\Users\AKILESH\downloads\DDOS.csv")
```

```python
data.dropna(inplace=True)
```

```python
# Import label encoder 
from sklearn import preprocessing 
  
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
```

```python
# Encode labels in column 'PKT_TYPE'. 
data['PKT_TYPE']= label_encoder.fit_transform(data['PKT_TYPE']) 
  
data['PKT_TYPE'].unique() 
```

```python
# Encode labels in column 'FLAGS'. 
data['FLAGS']= label_encoder.fit_transform(data['FLAGS']) 
  
data['FLAGS'].unique()
```

```python
# Encode labels in column 'NODE_NAME_FROM'. 
data['NODE_NAME_FROM']= label_encoder.fit_transform(data['NODE_NAME_FROM']) 
  
data['NODE_NAME_FROM'].unique()
```

```python
# Encode labels in column 'NODE_NAME_TO'. 
data['NODE_NAME_TO']= label_encoder.fit_transform(data['NODE_NAME_TO']) 
  
data['NODE_NAME_TO'].unique()
```

```python
# Encode labels in column 'PKT_CLASS'. 
data['PKT_CLASS']= label_encoder.fit_transform(data['PKT_CLASS']) 
  
data['PKT_CLASS'].unique() 
```

```python
# Now create an X variable (containing the features) and an y variable (containing only the target variable)
X = data.drop(["PKT_CLASS"], axis=1)
y = data['PKT_CLASS']

# Plot histograms of each parameter 
X.hist(figsize = (20, 20))
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X)

# Plot histograms of each parameter 
X.hist(figsize = (20, 20))
plt.show()
```

```python
#Split the data set using 'train_test_split' function
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_scaled = scaler.fit_transform(X_train)
test_scaled = scaler.transform(X_test)
from sklearn.neural_network import MLPClassifier
```

```python
model = MLPClassifier()
```

```python
model.fit(train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, model.predict(test_scaled))
```
